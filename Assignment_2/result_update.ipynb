```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as stats
```

```
# reading csv
df = pd.read_csv("mod_data.csv")
# Converting date column to datetime format
df['Date']= df['Date'].astype(str) 
for i in df['Date']:
    count = 0
    k = ""
    for j in i:
        count = count+1
        k = k+j
        if count == 4 or count == 6:
            k = k+"-"
    df['Date'][i] = k
df['Date']= pd.to_datetime(df['Date'])
```

#Scatterplot of C-rtn

```
df_C = df.filter(['Date','C'],axis=1)
ax = df_C.plot.scatter(x='Date', y='C')
```

#Scatterplot of SP-return

```
df_SP = df.filter(['Date','SP'],axis=1)
ax = df_SP.plot.scatter(x='Date', y='SP')
```

```
df_C_SP = df.filter(['C','SP'],axis=1)
var_s = df_C_SP.var()
mean_s = df_C_SP.mean()
```

#Normal Distribution for C-rtn

```
 plt.hist(np.array(df['C']),100, density=1, facecolor='g', alpha=0.5)
mu = mean_s['C']
variance = var_s['C']
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

#Normal distribution for SP return

```
plt.hist(np.array(df['SP']),100, density=True, facecolor='g', alpha=0.5)
mu = mean_s['SP']
variance = var_s['SP']
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```
kurtosis_C = stats.kurtosis(np.array(df['C']))
kurtosis_SP = stats.kurtosis(np.array(df['SP']))
skewness_C = stats.skew(np.array(df['C']))
skewness_SP = stats.skew(np.array(df['SP']))
print("kurtosis of C rtn = ", kurtosis_C)
print("kurtosis of SP rtn = ", kurtosis_SP)
print("skewness of C rtn = ", skewness_C)
print("skewness of SP rtn = ", skewness_SP)
```

#Double exponential distribution for C-rtn

```
plt.hist(np.array(df['C']),1000, density=True, facecolor='g', alpha=0.5)
mu = mean_s['C']
sigma = math.sqrt(var_s['C'])
x = np.linspace(mu -10*sigma, mu + 10*sigma, 1000)
g = stats.laplace.pdf(x,loc=mu,scale=sigma)
plt.plot(x,g)
plt.show()
```

#Double Exponential distribution for SP -rtn

```
plt.hist(np.array(df['SP']),1000, density=True, facecolor='g', alpha=0.5)
mu = mean_s['SP']
sigma = math.sqrt(var_s['SP'])
x = np.linspace(mu -10*sigma, mu + 10*sigma, 1000)
g = stats.laplace.pdf(x,loc=mu,scale=sigma)
plt.plot(x,g)
plt.show()
```

#t-distribution for C-rtn

```
plt.hist(np.array(df['C']),100, density=True, facecolor='g', alpha=0.5)
mu = mean_s['C']
sigma = math.sqrt(var_s['C'])

x = np.linspace(mu -10*sigma, mu + 10*sigma, 100)
g = stats.t.pdf(x,1,loc=mu,scale=sigma)
plt.plot(x,g)
plt.show()
```

# t-distribution for SP return

```
plt.hist(np.array(df['SP']),1000, density=True, facecolor='g', alpha=0.5)
mu = mean_s['SP']
sigma = math.sqrt(var_s['SP'])
x = np.linspace(mu - 10*sigma, mu + 10*sigma, 100)
g = stats.t.pdf(x,1,loc=mu,scale=sigma)
plt.plot(x,g)
plt.show()
```

# Cauchy distribution for C -rtn

```
plt.hist(np.array(df['C']),1000, density=True, facecolor='g', alpha=0.5)
mu = mean_s['C']
sigma = math.sqrt(var_s['C'])
x = np.linspace(mu -10*sigma, mu + 10*sigma, 100)
g = stats.cauchy.pdf(x,loc=mu,scale=sigma)
plt.plot(x,g)
plt.show()
```

#Cauchy distribution for SP return

```
plt.hist(np.array(df['SP']),1000, density=True, facecolor='g', alpha=0.5)
mu = mean_s['SP']
sigma = math.sqrt(var_s['SP'])
x = np.linspace(mu -10*sigma, mu + 10*sigma, 100)
g = stats.cauchy.pdf(x,loc=mu,scale=sigma)
plt.plot(x,g)
plt.show()
```

#Mix Normal distribution for C-rtn

```
plt.hist(np.array(df['C']),100, density=True, facecolor='g', alpha=0.5)
mu = mean_s['C']
sigma = 1
lambda_ = 0.2
mu1 = mu*(2/3)
mu2 = mu*(1/3)
x = np.linspace(mu -0.2, mu + 0.2, 100)
g = (lambda_)*stats.norm.pdf(x,mu1,0.1) + (1-lambda_)*stats.norm.pdf(x,mu2,0.01) 
plt.plot(x,g)
plt.show()
```

#Mix normal distribution for SP-rtn

```
plt.hist(np.array(df['SP']),1000, density=True, facecolor='g', alpha=0.5)
mu = mean_s['SP']
sigma = 1
lambda_ = 0.4
mu1 = mu*(2/3)
mu2 = mu*(1/3)
x = np.linspace(mu -0.2, mu + 0.2, 100)
g = (lambda_)*stats.norm.pdf(x,mu1,0.001) + (1-lambda_)*stats.norm.pdf(x,mu2,0.01) 
plt.plot(x,g)
plt.show()
```
